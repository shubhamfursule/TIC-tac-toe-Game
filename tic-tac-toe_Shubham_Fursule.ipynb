```python
from random import choice
def print_tic(frame):
    for i in frame: 
        print(*i)
def edit_tic_tac_toe(a,row1, column1):
    global frame
    if frame[row1][column1]==' ':
        frame[row1].pop(column1)
        frame[row1].insert(column1,a)
    
    print_tic(frame)  
    return frame
def row_column_convert(row11,column11):
    if row11==1: 
        row11=0
    elif row11==3:
        row11=4
    if column11==1:  
        column11=0
    elif column11==3:
        column11=4
    return row11,column11

def check_tic_tac_toe(frame1,sign):
    flag=False
    a=[0,2,4]
    for i in a:
        if frame1[i].count(sign)==3:
            #print(True)
            flag=True
            break

    if flag==False:
        diagonal=[]
        for i1 in a:
            k=[]
            for j1 in a:
                k.append(frame1[j1][i1])
                if i1==j1:
                    diagonal.append(frame1[i1][j1])
            if k.count(sign)==3:
                #print(True)
                flag=True
                break
        if diagonal.count(sign)==3:
            flag=True
            #print(True)
    if flag==False:
        
        diagonal2=[]
        diagonal2.append(frame1[0][4])
        diagonal2.append(frame1[2][2])
        diagonal2.append(frame1[4][0])
        #print(diagonal2)
        if diagonal2.count(sign)==3:
            flag=True
            #print(True)
    if flag==True:
        return True
    else:
        return False

def dict_tic(name,score):
    global Game_score
    
    if name not in Game_score:
        Game_score.update({name:score})
    else :
        Game_score[name]+=score
    return Game_score

Game_score={}
frame1=[]
def frame_tic():
    for i in range(5):
        list2=[]
        for j in range(5):
            if j%2==1 and i%2==1:
                list2.append('|')
            elif i%2==0 and j%2==0:            
                list2.append(" ")
            elif i%2==1 and j%2==0:
                list2.append('_')
            elif i%2==0 and j%2==1:
                list2.append('|')
        frame1.append(list2)
    return frame1
frame=frame_tic()
print_tic(frame)
print("!!Please Select the Mode for Playing Tic-Tac-Toe!!")
print("1: Single Player \n2: Two Players ")
try:
    Mode=int(input("Enter Mode : "))
    if Mode>2 or Mode<1:
        raise "Error"
except:
    if "Error":
        print("Enter properly!!")
else:
    if Mode==1:
        Player_name=input("Enter Player name: ")
        count1=0
        check_t_X_1={}
        while count1!=9:
            if count1%2==0:
                try:
                    row_1,column_1=map(int,input("Enter row and column: ").split())
                    if row_1<1 or row_1>3 or column_1 <1 or column_1>3 :
                        raise "ValueError_1"
                except:
                    if "ValueError_1":
                        print("Enter value in a proper way !")
                else:
                    row_new_1,column_new_1=row_column_convert(row_1,column_1)
                    if frame[row_new_1][column_new_1]==' ':
                        a= edit_tic_tac_toe('X',row_new_1,column_new_1)
                        check_t_X_1=check_tic_tac_toe(frame,'X')
                        if check_t_X_1==True:
                            print(f"{Player_name} is Win ")
                            print("Score card")
                            score_ran=dict_tic(Player_name,10)
                            for x,y in score_ran.items():
                                print(x,y)
                            break
                        else:
                            count1+=1
                    else:
                        print("!! That position is allready field, Try another one  !!")
                        continue
            else:
                flag_ran=False
                
                list_index_g=[0,2,4]
                row_random=choice(list_index_g)
                column_random=choice(list_index_g)
                if frame[row_random][column_random]==' ':
                    print("computer :\n ")
                    b=edit_tic_tac_toe('O',row_random,column_random)
                    count1+=1
                    
                else:
                    continue
                    
                check_O_ran=check_tic_tac_toe(frame,'O')
                if check_O_ran==True:
                    print(f"{Player_name} is loss game ")
                    flag_ran=True
                    break
                
                                    
    elif Mode==2:
        first_player=input("Name of 1st Player: ")
        Second_player=input("Name of 2nd Player : ")
        count=0
        check_t_X={}
        check_t_O={}
        while True:
            while count!=9:
                try:
                    if count%2==0:
                        try:
                            row,column=map(int,input(f"'{first_player}' Enter row and column : ").split())
                            if row<1 or row>3 or column <1 or column>3 :
                                raise "ValueError1"
                        except:
                            if "ValueError1":
                                print("Enter value in a proper way!")
                        else:
                            row_new,column_new=row_column_convert(row,column)
                            if frame[row_new][column_new]==' ':
                                a= edit_tic_tac_toe('X',row_new,column_new)
                                check_t_X=check_tic_tac_toe(frame,'X')
                                
                                if check_t_X==True:
                                    print(f"{first_player} is Win ")
                                    print("Score card")
                                    score=dict_tic(first_player,10)
                                    for x,y in score.items():
                                        print(x,y)
                                    break
                                else:
                                    count+=1
                            else:
                                print("!! That position is allready field, Try another one  !!")
                                continue
                    else:
                        try:
                            row,column=map(int,input(f"'{Second_player}' Enter row and column : ").split())
                            if row<1 or row>3 or column <1 or column>3 :
                                raise "ValueError11 "
                        except :
                            print("Enter value in a proper way!")
                        else:
                            row_new,column_new=row_column_convert(row,column)
                            if frame[row_new][column_new]==' ':
                                a= edit_tic_tac_toe('O',row_new,column_new)
                                check_t_O=check_tic_tac_toe(frame,'O')
                                if check_t_O==True:
                                    print(f"{Second_player} is Win ")
                                    score1=dict_tic(Second_player,10)
                                    for x,y in score1.items():
                                        print(x,y)
                                    break
                                else:
                                    count+=1
                            else:
                                print("!! That position is allready field, Try another one  !!")
                                continue

                    if check_t_X==False and check_t_O==False and count==9:
                        print("\n!! Game is tie :( !!")
                        print("Game over :o ")
                        score2=dict_tic(first_player,5)
                        score2=dict_tic(Second_player,5)
                        for x,y in score2.items():
                            print(x,y)   

                except:
                    print("valid!")
            print("\n1. Play Again  \n2: Do You want Quit Game")
            Quit_not=int(input("\nEnter above option : "))
            if Quit_not==1:
                frame1=[]
                frame=frame_tic()
                print_tic(frame)
                count=0
                continue
            else:                
                break
        l=list(Game_score.keys())
        if Game_score[l[0]]==Game_score[l[1]]:
            print(f"\n Game is tie ")
        elif Game_score[l[0]]>Game_score[l[1]]:
            print(f"\n{l[0]} is win")
        else:
            print(f"\n{l[1]} is win")



```
